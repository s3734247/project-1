```python
iris_filename = 'Automobile.csv'
```

```python
iris=pd.read_csv(iris_filename,sep='#',decimal='.',header=None,names=['symboling','normalized-losses','make','fuel-type','aspiration','num-of-doors','body-style','drive-wheels','engine-location','wheel-base','length','width','height','curb-weight','engine-type','num-of-cylinders','engine-size','fuel-system','bore','stroke','compression-ratio','horsepower','peak-rpm','city-mpg','highway-mpg','price'],error_bad_lines=False)
```

```python
iris=iris.replace('vol00112ov','volvo')
```

```python
iris=iris.fillna(iris.mean(axis=0).map(lambda x:('%.2f')%x) )
```

```python
iris['normalized-losses'].describe()
```

```python
iris=iris[~iris['symboling'].isin([4])]
```

```python
iris=iris.replace('turrrrbo','turbo')
```

```python
iris=iris.replace('fourR','four')
```

```python
iris['num-of-cylinders'] = iris['num-of-cylinders'].apply(lambda x: x.lower())
```

```python
iris['engine-location'] = iris['engine-location'].apply(lambda x: x.lower())
```

```python
iris['fuel-type'] = iris['fuel-type'].apply(lambda x: x.lower())
```

```python
iris['aspiration'] = iris['aspiration'].apply(lambda x: x.lower())
```

```python
iris['body-style'] = iris['body-style'].apply(lambda x: x.lower())
```

```python
iris.describe()
```

```python
iris=iris.replace(' ','')
```

```python
iris['make']=iris['make'].str.strip();
```

```python
iris['aspiration']=iris['aspiration'].str.strip(); 
```

```python
iris['engine-location']=iris['engine-location'].str.strip(); 
```

```python
iris['fuel-type']=iris['fuel-type'].str.strip(); 
```

```python
iris['body-style']=iris['body-style'].str.strip(); 
```

```python
iris=iris.dropna(axis = 1)
```
